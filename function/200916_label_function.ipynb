```python
import pandas as pd
from IPython.display import clear_output
```

```python
df = pd.read_csv(r'D:\local\insta_profiling\data\jl_id_region_2.csv')
```

```python
def label(df, start_idx=0):
    for x in range(start_idx, len(df)):
        print('현재 지역 라벨: ', df.loc[x][0])
        print([location for location in df.loc[x] if type(location)==str][1:])
        region_label = input('사용자의 지역을 숫자로 입력해 주세요(종료를 원하시면 0을 입력해 주세요)\n수도권(1), 강원(2), 충청(3), 전라(4), 경상(5), 모름(6) ')
        if region_label == '0':
            print('종료 인덱스: ', x)
            break
        df['region'][x] = region_label
        clear_output()
        if x == len(df)-1:
            print('더 이상 라벨링할 데이터가 없습니다!')
```

```python
# (데이터프레임, 라벨링 시작하고 싶은 인덱스)를 함수에 넣어서 실행
label(df, 347)
```

```python
df.to_csv(r'D:\local\insta_profiling\data\jl_labeling.csv', index=False)
```

```python
import os
os.chdir(r'D:\local\insta_profiling\data\labeling')
```

```python
lst = os.listdir()
df = pd.read_csv(lst[0])
```

```python
new_df = df[['region', 'insta_id']]
for x in lst[1:]:
    df2 = pd.read_csv(x)
    new_df2 = df2[['region', 'insta_id']]
    new_df = pd.concat([new_df, new_df2])
```

```python
def num_to_region(num):
    if num == 6.0:
        return 'unk'
    elif num == 5.0:
        return '경상'
    elif num == 4.0:
        return '전라'
    elif num == 3.0:
        return '충청'
    elif num == 2.0:
        return '강원'
    else:
        return '수도'
```

```python
new_df['region'] = new_df['region'].apply(num_to_region)
```

```python
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df = new_df[new_df['region']!='unk']
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df.to_csv('label_result.csv', index=False)
```
