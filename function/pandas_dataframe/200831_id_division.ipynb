```python
import pandas as pd
import os
import re
```

```python
def id_division(csv_name):
    df = pd.read_csv(csv_name)
    lst = [(x, y) for x, y in zip(df['inner_id'], df['insta_id'])]
    new_lst = list(set(lst))
    dic = {}
    for idx, x in enumerate(new_lst):
        dic_num = idx//500 + 1
        if idx%500 == 0:
            dic['{}'.format(dic_num)] = [x]
        else:
            dic['{}'.format(dic_num)].append(x)
    for df_num, data in dic.items():
        new_df = pd.DataFrame(data, columns = ['inner_id', 'insta_id'])
        new_df.to_csv(csv_name[:-4]+'_'+df_num+'.csv', index=False)
        
```

```python
os.chdir('D:\local')
df = pd.read_csv('jl_word.csv')
```

```python
def post_division(df, n, csv_name):
    dic_id = {}
    id_lst = list(set(df['insta_id']))
    print(len(id_lst))
    for i, id in enumerate(id_lst):
        dic_num = i//n
        if i%n == 0:
            dic_id['{}'.format(dic_num)] = [id]
        else:
            dic_id['{}'.format(dic_num)].append(id)
    for num, ids in dic_id.items():
        new_df = pd.DataFrame(columns = df.columns)
        for id in ids:
            new_df = new_df.append(df[df['insta_id']==id])
        new_df.to_csv('{}_{}.csv'.format(csv_name, int(num)+1), index=False)
        
        
```
